```python
# Import stuff
import numpy as np
import matplotlib.pyplot as plt
from array import array
# Mass & radius of sun
msun= 2E30
rsun = 696340000
# Calculate central density and pressure for various radii of a 0.03 Msun star
mass = 0.03*msun
radius = np.linspace(0.01*rsun,10*rsun,1000)

rho_c_003 = 5.99*mass/(4/3*np.pi*radius**3)
P_c = 0.77*6.67E-11 * mass**2 / radius**4
# Calculate P degenerate and other stuff
P_deg = 9.91E12 * (rho_c_003/1000)**(5/3) / 1.176**(5/3) / 10
x_deg = P_deg / P_c
fe = 0.525
# Convert arrays to a list
x_deg = list(x_deg)
P_c = list(P_c)
rho_c_003 = list(rho_c_003)

# Delete list indices that don't make sense
indices = 0
for P in range(len(x_deg)):
    if x_deg[P]>1:
        indices += 1

x_deg = x_deg[indices:]
P_c = P_c[indices:]
rho_c_003 = rho_c_003[indices:]

# Convert back into array
x_deg_003 = np.array(x_deg)
P_c = np.array(P_c)
rho_c_003 = np.array(rho_c_003)
# Find core temperature
x_ideal_003 = (1-x_deg_003**2) / (1-fe + np.sqrt(fe**2+ (1-2*fe)*x_deg_003**2))
T_c_003 = x_ideal_003*P_c*0.617*1.6735E-27/(rho_c_003*1.3806E-23)
```

```python
# Repeat with a 0.1 Msun star
mass = 0.1*msun

rho_c_01 = 5.99*mass/(4/3*np.pi*radius**3)
P_c = 0.77*6.67E-11 * mass**2 / radius**4

P_deg = 9.91E12 * (rho_c_01/1000)**(5/3) / 1.176**(5/3) / 10
x_deg = P_deg / P_c
fe = 0.5247

x_deg = list(x_deg)
P_c = list(P_c)
rho_c_01 = list(rho_c_01)

P_array = []
for P in range(len(x_deg)):
    if x_deg[P]>1:
        P_array.append(P)

x_deg = x_deg[len(P_array):]
P_c = P_c[len(P_array):]
rho_c_01 = rho_c_01[len(P_array):]

x_deg_01 = np.array(x_deg)
P_c = np.array(P_c)
rho_c_01 = np.array(rho_c_01)

x_ideal_01 = (1-x_deg_01**2) / (1-fe + np.sqrt(fe**2+ (1-2*fe)*x_deg_01**2))
T_c_01 = x_ideal_01*P_c*0.617*1.6735E-27/(rho_c_01*1.3806E-23)
```

```python
# Repeat with a 0.3 Msun star
mass = 0.3*msun

rho_c_03 = 5.99*mass/(4/3*np.pi*radius**3)
P_c = 0.77*6.67E-11 * mass**2 / radius**4

P_deg = 9.91E12 * (rho_c_03/1000)**(5/3) / 1.176**(5/3) / 10
x_deg = P_deg / P_c
fe = 0.5247

x_deg = list(x_deg)
P_c = list(P_c)
rho_c_03 = list(rho_c_03)

P_array = []
for P in range(len(x_deg)):
    if x_deg[P]>1:
        P_array.append(P)
x_deg = x_deg[len(P_array):]
P_c = P_c[len(P_array):]
rho_c_03 = rho_c_03[len(P_array):]

x_deg_03 = np.array(x_deg)
P_c = np.array(P_c)
rho_c_03 = np.array(rho_c_03)

x_ideal_03 = (1-x_deg_03**2) / (1-fe + np.sqrt(fe**2+ (1-2*fe)*x_deg_03**2))
T_c_03 = x_ideal_03*P_c*0.617*1.6735E-27/(rho_c_03*1.3806E-23)
```

```python
# Repeat with a 1 Msun star
mass = msun

rho_c_1 = 5.99*mass/(4/3*np.pi*radius**3)
P_c = 0.77*6.67E-11 * mass**2 / radius**4

P_deg = 9.91E12 * (rho_c_1/1000)**(5/3) / 1.176**(5/3) / 10
x_deg = P_deg / P_c
fe = 0.5247

x_deg = list(x_deg)
P_c = list(P_c)
rho_c_1 = list(rho_c_1)

P_array = []
for P in range(len(x_deg)):
    if x_deg[P]>1:
        P_array.append(P)

x_deg = x_deg[len(P_array):]
P_c = P_c[len(P_array):]
rho_c_1 = rho_c_1[len(P_array):]

x_deg_1 = np.array(x_deg)
P_c = np.array(P_c)
rho_c_1 = np.array(rho_c_1)

x_ideal_1 = (1-x_deg_1**2) / (1-fe + np.sqrt(fe**2+ (1-2*fe)*x_deg_1**2))
T_c_1 = x_ideal_1*P_c*0.617*1.6735E-27/(rho_c_1*1.3806E-23)
```

```python
# Plot the stuff, also convert to cringe cgs units
plt.scatter(rho_c_003/1000, T_c_003, label='0.03 Solar M')
plt.scatter(rho_c_01/1000, T_c_01, label='0.1 Solar M')
plt.scatter(rho_c_03/1000, T_c_03, label='0.3 Solar M')
plt.scatter(rho_c_1/1000, T_c_1, label='1 Solar M')
plt.xscale("log")
plt.yscale("log")
plt.xlabel("Central density (g/cm^3)")
plt.ylabel('Central temperature (K)')
plt.legend(loc='upper left')
plt.xlim(10E-13,10E7)
plt.ylim(10E2,10E7)
plt.show()
```

```python
# Temperature-mass relation
T_max=[np.max(T_c_003),np.max(T_c_01),np.max(T_c_03),np.max(T_c_1)]
masses = [0.03,0.1,0.3,1]
plt.scatter(np.log10(masses), np.log10(T_max))
plt.plot([-1.52,0],[5.96,7.99],linestyle='--')
plt.xlabel('log mass (Solar masses)')
plt.ylabel('log temperature (K)')
plt.title("Temperature-Mass Profile")
plt.show()
```

```python
# Unlike the other arrays radius hasn't been cut down yet so we're just gonna do it manually here
r_max=[radius[np.argmax(T_c_003)+9]/696340000,radius[np.argmax(T_c_01)+6]/696340000,
       radius[np.argmax(T_c_03)+4]/696340000,radius[np.argmax(T_c_1)+3]/696340000]
plt.scatter(np.log10(masses),np.log10(r_max))
plt.plot([-1.52,0],[-0.78,-1.28],linestyle='--')
plt.xlabel('log mass (solar masses)')
plt.ylabel('log radius (solar radii)')
plt.title("Radius-Mass Profile")
plt.show()
```

```python
densities = [rho_c_003[np.argmax(T_c_003)]/1000,rho_c_01[np.argmax(T_c_01)]/1000,rho_c_03[np.argmax(T_c_03)]/1000,
             rho_c_1[np.argmax(T_c_1)]/1000]
plt.scatter(np.log10(masses),np.log10(densities))
plt.plot([-1.52,0],[1.73,4.77],linestyle='--')
plt.xlabel('log mass (solar masses)')
plt.ylabel('log density (g/cm^3)')
plt.title("Density-Mass Profile")
plt.show()
```
